```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
from tensorwaves.transfer import PhaseAberration, CTF, TemporalEnvelope
```

```python
aberrations = PhaseAberration(energy=200e3)
aberrations.parametrization.defocus = -600
aberrations.parametrization.Cs = 1e7

k, phi = aberrations.line_data(0, k_max=.8, n=256)

plt.plot(k, phi.numpy().imag)

aberrations.parametrization.defocus = -660
k, phi = aberrations.line_data(0, k_max=.8, n=256)

plt.plot(k, phi.numpy().imag)
plt.ylabel('MTF')
plt.xlabel('Spatial Frequency (in 1/Ã…)')
```

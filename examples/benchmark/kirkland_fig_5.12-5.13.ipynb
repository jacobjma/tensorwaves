# Kirkland Fig. 5.12 & 5.13

```python
import numpy as np
from ase import Atoms
import matplotlib.pyplot as plt

from tensorwaves.waves import PlaneWaves
from tensorwaves.potentials import Potential
```

```python
atoms = Atoms('CSiCuAuU', positions=[(x, 25, 5) for x in np.linspace(5, 45, 5)], cell=(50, 50, 10))

potential = Potential(atoms=atoms, gpts=512, parametrization='kirkland')

wave = PlaneWaves(energy=200e3)

wave = wave.multislice(potential)

wave = wave.apply_ctf(defocus=700, Cs=-1.3e7, aperture_cutoff=.01037)

image = wave.image()

plt.plot(image.numpy()[0,:,256])
plt.ylabel('Image Intensity')
plt.xlabel('position x (in Ang)')
plt.ylim([.5,1.1])
```

```python
%load_ext autoreload
%autoreload 2
import numpy as np
from ase.build import mx2
from tensorwaves.waves import PlaneWaves
from tensorwaves.potentials import Potential
import matplotlib.pyplot as plt
```

```python
atoms = mx2(formula='MoS2', kind='2H', a=3.18, thickness=3.19)
atoms *= (2,2,1)
atoms.cell[1,0] = 0
atoms.wrap()
atoms *= (2,2,1)
atoms.center(vacuum=2, axis=2)
```

```python
waves = PlaneWaves(sampling=.05, energy=80e3)
waves = waves.multislice(atoms)
```

```python
waves = waves.apply_ctf(defocus=200, aperture_radius=1)
```

```python
plt.imshow(waves.image().numpy()[0])
```
